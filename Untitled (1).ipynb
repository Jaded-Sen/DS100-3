```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('df.csv', encoding='latin1', skipfooter=5).drop(['Series Code', 'Country Name'], axis=1)
df.replace('..', np.nan, inplace=True)
df.columns = df.columns.str.replace('\[YR\d{4}\]', 'IT', regex=True)
df = df[df['Series'] == 'Internet users (per 100 people)']
df[['2020 IT', '2019 IT', '2018 IT', '2017 IT', '2016 IT', '2015 IT', '2014 IT', '2013 IT', '2012 IT', '2011 IT', '2010 IT']] = df[['2020 IT', '2019 IT', '2018 IT', '2017 IT', '2016 IT', '2015 IT', '2014 IT', '2013 IT', '2012 IT', '2011 IT', '2010 IT']].astype(float)
df = df.set_index('Country Code').drop(['ANT', 'FTI', 'LNX', 'MYT', 'TKL']).sort_index(ascending=True).drop('Series', axis=1)
```

```python
df_gdp = pd.read_csv('gdp.csv', skipfooter=5).drop(['Series Name', 'Series Code', 'Country Name'], axis=1)
df_gdp.replace('..', np.nan, inplace=True)
df_gdp.columns = df_gdp.columns.str.replace('\[YR\d{4}\]', 'GDP', regex=True)
df_gdp[['2020 GDP', '2019 GDP', '2018 GDP', '2017 GDP', '2016 GDP', '2015 GDP', '2014 GDP', '2013 GDP', '2012 GDP', '2011 GDP', '2010 GDP']] = df_gdp[['2020 GDP', '2019 GDP', '2018 GDP', '2017 GDP', '2016 GDP', '2015 GDP', '2014 GDP', '2013 GDP', '2012 GDP', '2011 GDP', '2010 GDP']].astype(float)
df_gdp = df_gdp.set_index('Country Code').drop(['AFE', 'AFW', 'INX']).sort_index(ascending=True)
```

```python
plt.figure(figsize=(15,8))

x = np.linspace(0, 100, 100)
y = np.exp(x/8.2)

plt.title('GDP per capita vs Internet Access')
plt.xlabel('Internet users (per 100 people)')
plt.ylabel('GDP per capita')

plt.scatter(df, df_gdp, label='GDP per capita vs Internet Access')
plt.plot(x, y, c='red', label='Exponential Function')
plt.legend(loc=2)

for spine in plt.gca().spines.values():
    spine.set_visible(False)
```

```python
x_axis = df.columns.tolist()[::-1]
plt.figure(figsize=(20,12))

plt.title('GDP per capita evolution - South America')
plt.ylabel('GDP per capita')
plt.xlabel('Year')

plt.plot(x_axis, df_gdp.loc['ARG'][::-1], c='blue', label='ARG')
plt.plot(x_axis, df_gdp.loc['CHL'][::-1], c='red', label='CHL')
plt.plot(x_axis, df_gdp.loc['BRA'][::-1], c='yellow', label='BRA')
plt.plot(x_axis, df_gdp.loc['ECU'][::-1], c='black', label='ECU')
plt.plot(x_axis, df_gdp.loc['BOL'][::-1], c='green', label='BOL')
plt.plot(x_axis, df_gdp.loc['COL'][::-1], c='magenta', label='COL')
plt.plot(x_axis, df_gdp.loc['PER'][::-1], c='cyan', label='PER')
plt.plot(x_axis, df_gdp.loc['URY'][::-1], c='gray', label='URY')
plt.plot(x_axis, df_gdp.loc['PRY'][::-1], c='orange', label='PRY')
plt.plot(x_axis, df_gdp.loc['VEN'][::-1], c='purple', label='VEN')
plt.legend()
```
