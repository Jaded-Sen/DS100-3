```python
import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
from scipy import stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

ave=df.mean(axis=1)
std=df.std(axis=1)
yerr=std/np.sqrt(df.shape[1])*stats.t.ppf(0.95, df.shape[1]-1)
plt.figure()
plt.show()


bars=plt.bar(range(df.shape[0]),ave,yerr=yerr, alpha=0.6, color=['red','green','blue','yellow'])

threshold=42000
plt.axhline(y=threshold, color='gray', alpha=1)
plt.xticks(range(df.shape[0]), ['1992', '1993', '1994', '1995'], alpha=0.8)
plt.title('Ferreira et al, 2014: Harder option')
plt.xlabel('Year')
plt.ylabel('Values')           
```
