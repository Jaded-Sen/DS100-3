IMPORTING DATA AND CLEANING

```python
import pandas as pd

shaq = pd.read_csv('https://raw.githubusercontent.com/Jaded-Sen/DS100-3/main/shaq-nba-career-regular-season-stats-by-game.csv')

shaq.head()
```

```python
shaq = shaq.iloc[1:]
shaq = shaq[['Years', 'Win']]
shaq.dropna(inplace=True)
shaq.head()
```

```python
shaq.rename(columns={'Years': 'Age'}, inplace = True)

shaq_group = shaq.groupby('Age')
shaq2 = shaq_group.sum().reset_index()
shaq2

import matplotlib.pyplot as plt

plt.plot(shaq2.Age, shaq2.Win)
plt.show
```

```python
shaq3 = shaq.groupby('Age', as_index=True).sum().reset_index()

shaq3
```

```python
mic = pd.read_csv('https://raw.githubusercontent.com/Jaded-Sen/DS100-3/main/michael-jordan-nba-career-regular-season-stats-by-game.csv')

mic.head()
```

```python
mic = mic.iloc[1:]
mic = mic[['Years', 'Win']]
mic.dropna(inplace=True)
mic.head()
```

```python
mic.rename(columns={'Years': 'Age'}, inplace = True)
mic
mic_group = mic.groupby('Age')
mic2 = mic_group.sum().reset_index()
mic2

shaq.rename(columns={'Years': 'Age'}, inplace = True)

shaq_group = shaq.groupby('Age')
shaq2 = shaq_group.sum().reset_index()
shaq2

import matplotlib.pyplot as plt
import numpy as np

plt.figure(num=None, figsize=(14,10), dpi=80)

plt.plot(mic2.Age, mic2.Win, color='red',  label = 'Jordan', alpha = 0.5, linewidth = 3, marker='o', markerfacecolor='azure', markeredgecolor='dimgray', markeredgewidth='3')
plt.plot(shaq2.Age, shaq2.Win, color='blue',  label = 'Shaq', alpha = 0.5, linewidth = 3, marker='o', markerfacecolor='lightsteelblue', markeredgecolor='dimgray', markeredgewidth='3')

plt.xticks(np.arange(20, 41, 1), np.arange(20, 41, 1), fontsize=10)
plt.yticks(np.arange(10, 74, 2), np.arange(10, 74, 2), fontsize=10)

legend = plt.legend(loc='upper right', shadow=False, frameon=True, fontsize=15)

plt.ylabel('Win', fontweight='bold', fontsize='15')
plt.xlabel('Age', fontweight='bold', fontsize='15')
plt.title("Michael Jordan and Shaq O'Niel Number of Wins by Age", fontweight='bold', fontsize='15')

plt.show
```

```python
mic3 = mic.groupby('Age', as_index=True).sum().reset_index()

mic3
```

```python
both = shaq2.merge(mic2, left_on='Age', right_index=True, suffixes=('_Shaq', '_Jordan'))
both

plt.plot(both.Age, both.Win_Shaq, color='red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


plt.figure(num=None, figsize=(14,10), dpi=80)

plt.plot(shaq2.Age, shaq2.Win, color = 'red', label = 'Shaq', alpha = 0.5, linewidth = 5)
plt.plot(mic2.Age, mic2.Win, color = 'blue', label = 'Jordan', alpha = 0.5, linewidth = 5)
plt.ylim(0,1)

plt.ylabel('Number of Win')
plt.xlabel('Age')
plt.title("Michael Jordan and Shaq O'Niel Number of Wins by Age")

legend = plt.legend(loc='upper right', shadow=False, frameon=True, fontsize=15)

for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.grid(False)

plt.grid(linestyle = 'dashed', linewidth=0.5, axis='y')

sns.set_style("whitegrid")

plt.show()
```
